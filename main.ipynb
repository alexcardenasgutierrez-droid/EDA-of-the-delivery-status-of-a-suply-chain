## <h2 style="color: #31b0d6;">Defining some open-ended questions</h2>

After inspectioning the columns, we defined two open-ended questions:
- **What could affect the shipment time?**
- **What is the impact of the origin of the order and the quantity of items per order on the Delivery status**

For that we chose to keep the columns that could be related with the shipment and create an new data frame called sf


```python
# manually decided which column to keep -> put them in a list -> copy only these columns from the data frame
req_col_delivery_time = ['Days for shipping (real)', 'Days for shipment (scheduled)', 'Sales', 'Order Item Discount', 'Delivery Status', 'Market', 'Order Country', 'Order Id', 'Order Item Quantity']
sf = df[req_col_delivery_time].copy()
sf = sf.sort_values("Order Id")
sf.head()
```

## <h2 style="color: #31b0d6;">Data cleaning</h2>

- **We want to group by 'order_id' all the table**

```python
sf.columns
```

```python
# We count with nunique() the number of unique values for each column grouped by "order_id"
# if the group has less than 1 value the condition is True, and with .all we check if all groups meet the condition
print((sf.groupby("Order Id")["Days for shipping (real)"].nunique() <= 1).all())
print((sf.groupby("Order Id")["Days for shipment (scheduled)"].nunique() <= 1).all())
print((sf.groupby("Order Id")["Delivery Status"].nunique() <= 1).all())
print((sf.groupby("Order Id")["Market"].nunique() <= 1).all())
print((sf.groupby("Order Id")["Order Country"].nunique() <= 1).all())
```

```python
sf = sf.groupby("Order Id").agg({      
        'Days for shipping (real)':"first",
        'Days for shipment (scheduled)': "first",
        'Delivery Status': "first",
        'Market': "first",
        'Order Country': "first",
        'Order Item Quantity': "sum",

        'Sales': "sum",
        'Order Item Discount': "sum",
        })

# calculate the discount ratio
sf["discount_ratio"] = sf["Order Item Discount"]/sf["Sales"]

# drop the not neede columns after calculating the discount ratio
sf = sf.drop(columns=["Sales", "Order Item Discount"])

# rename columns (only lower case and change " " to "_")
sf.columns = sf.columns.str.lower().str.replace(" ", "_").str.replace(r"[()]", "", regex=True)

display(sf)

df["Order Id"].nunique()                    # -> checking of the count of unique Order Ids is equal to the rows after grouping
```
